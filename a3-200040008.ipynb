# Streamlines $\tau$ = 1

```python
x, y = np.mgrid[-2:2:100j, -2:2:100j]
u_inf = 1
r = 1
tau = 1
z = x + 1j*y
phi = u_inf*(z + (r**2)/z) + (-1j*tau*np.log(z))/2*np.pi
```

# Pressure Field $\tau$ = 1

```python
x, y = np.mgrid[-2:2:20j, -2:2:20j]
u_inf = 1
r = 1
p_inf = 1
rho = 1
tau = 1
z = x + 1j*y

v =  u_inf*(1-((r)**2)/(z*z)) -1j*tau/(2*np.pi*z)

p = p_inf + 0.5*rho*(u_inf - (v.real**2 + v.imag**2))
```

```python
plt.contour(x, y, p, levels = 5000,origin='image')
#Plot for pressure contours , note that it is dense at the centre since contours are closely placed
```

```python
plt.imshow(p,cmap='hot');
plt.colorbar();
#Heat map for Pressure field, the colour is very light since P -> P_inf-1/2*rho(U_inf^2 - u^2)as we increase distance from 
#source and approaches 0
```

# Drag & Lift

```python
rho = 1
u_inf = 1
tau = 1
r = 1
p_inf = 1

#ps = 0.5*rho*(u_inf**2 - (-2*u_inf*np.sin() + tau/(2*np.pi*r))**2 + p_inf)
```

```python
x = np.arange(0,2*np.pi,0.01)
dtheta = 0.01

drag = 0
lift = 0
for theta in x :
    theta = theta - dtheta/2
    ps = 0.5*rho*(u_inf**2 - (-2*u_inf*np.sin(theta) + tau/(2*np.pi*r))**2) + p_inf
    drag += ps*np.cos(theta)*r*dtheta
    lift += ps*np.sin(theta)*r*dtheta
```

```python
drag,lift
```

```python
# Calculated drag and lift as integration of {pressure x cos(theta) x r x d_theta} &
# integration of {pressure x sin(theta) x r x d_theta} respectively by taking small interval of 0.01 between successive 
# steps and taking the mid point of the step as input to pressure which is a function of theta on the surface of cylinder
```
