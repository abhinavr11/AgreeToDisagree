```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

# Part 1

## Pathlines 1

```python
def vel1(r,t):
    x,y = r
    #print(x,y)
    #X.append(x)
    #Y.append(y)
    if (x**2 + y**2)** 0.5 < 1:
        u = 0
        v = 0
    
    
   
    else :
        u = ((x**2 - y**2 - 1)*(x**2 - y**2) + 4* x**2 * y**2 )/((x**2 + y**2)**2)
        v = -2*x*y/((x**2 + y**2)**2)
    
    #print(u,v)
    #return np.array([u, v])
    return np.array([u, v])
```

```python


for i in range(10):
    ic = np.array([-2,-2 + i*0.45])
    times = np.linspace(0, 10, 100)

    px, py = odeint(vel1, ic, times).T
    plt.plot(px, py,'-o')
    plt.grid()
    #plt.axis('equal');
    
    #Plot for path lines for 10 particles on x=-2
```

## Stream lines 1

```python
def vel(x,y):
    x = x
    y = y
    
    if (x**2 + y**2)** 0.5 < 1:
        u = 0
        v = 0
    
    
   
    else :
        u = ((x**2 - y**2 - 1)*(x**2 - y**2) + 4* x**2 * y**2 )/((x**2 + y**2)**2)
        v = -2*x*y/((x**2 + y**2)**2)
    

    return u,v
```

```python
xo, yo = -2, -2
n = 25
dx = dy = 4./n
u = np.zeros((n, n))
v = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        u[i, j],v[i,j] = vel(xo + i*dx, yo+j*dy)
```

```python
x, y = np.mgrid[-1:1:25j, -1:1:25j]
plt.clf()
plt.quiver(x, y, u, v);

#Plot for velocity field 
```

## Streaklines 1


```python
def vel1(r,t):
    x,y = r

    if (x**2 + y**2)** 0.5 < 1:
        u = 0
        v = 0
    
    
   
    else :
        u = ((x**2 - y**2 - 1)*(x**2 - y**2) + 4* x**2 * y**2 )/((x**2 + y**2)**2)
        v = -2*x*y/((x**2 + y**2)**2)
    

    return np.array([u, v])


for j in range(10):
    ic = np.array([-2,-2 + j*0.45])

    dt = 0.1
    n = 100
    T = n*dt
    sx, sy = [], []
    for i in range(n):
        start = i*dt
        res = odeint(vel1, ic, [start, T])
        sx.append(res[1,0])
        sy.append(res[1,1])
    plt.plot(sx, sy)
    plt.grid()
    plt.axis('equal');
```

```python
#Above is Plot for streaklines for particles on x=-2
```
