```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def vel(x,y,q=1):
   
    
    if ((x+0.5)**2 + (y-0)**2)** 0.5 < 0.5 or ((x-0.5)**2 + (y-0)**2)** 0.5 < 0.5:
        u = 0
        v = 0
    
    
   
    else :
        up = q*(x+0.5)/((2*np.pi) * ((x+0.5)**2 + (y-0)**2)) 
        vp = q*y/((2*np.pi) * ((x+0.5)**2 + (y-0)**2)) 
        un = -q*(x-0.5)/((2*np.pi) * ((x-0.5)**2 + (y-0)**2)) 
        vn = -q*y/((2*np.pi) * ((x-0.5)**2 + (y-0)**2)) 
        
        u= up+un
        v= vp+vn

    return u,v
```

```python
xo, yo = -2, -2
n = 25
dx = dy = 4./n
u = np.zeros((n, n))
v = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        u[i, j],v[i,j] = vel(xo + i*dx, yo+j*dy)
```

```python
x, y = np.mgrid[-2:2:25j, -2:2:25j]
plt.clf()
plt.quiver(x, y, u, v);

#Plot for velocity field 
```

# Potential Lines

```python
def pot(x,y,q=1):
   
    if ((x+0.5)**2 + (y-0)**2)** 0.5 < 0.5 or ((x-0.5)**2 + (y-0)**2)** 0.5 < 0.5:
        phi_src = 0
        phi_sink = 0
    
    
   
    else :
        phi_src = q* np.log(((x+0.5)**2 + y**2)**0.5)
        phi_sink = -q* np.log(((x-0.5)**2 + y**2)**0.5)
      
    phi = phi_src+phi_sink

    return phi
```

```python
xo, yo = -3, -3
n = 50
dx = dy = 6./n
p = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        
        p[i,j] = pot(xo + i*dx, yo+j*dy)

```

```python
plt.imshow(p,cmap='hot');
plt.colorbar();
plt.contour(p,levels=100);

#Plot for Potential function along with contours (Phi is set to 0 for r < 0.5)
```

# Adding freestream

```python
def pot(x,y,q=1):
   
    if ((x+0.5)**2 + (y-0)**2)** 0.5 < 0.05 or ((x-0.5)**2 + (y-0)**2)** 0.5 < 0.05 or x ==0:
        phi_src = 0
        phi_sink = 0
    
    
   
    else :
        phi_src = 1*np.cos(np.arctan(y/x)) + q* np.log(((x+0.5)**2 + y**2)**0.5)
        phi_sink = -q* np.log(((x-0.5)**2 + y**2)**0.5)
      
    phi = phi_src+phi_sink

    return phi
```

# Pressure field

```python
def pressure(x,y,q=1):
    epsilon = 0.01
    
    if ((x+0.5)**2 + (y-0)**2)** 0.5 < epsilon or ((x-0.5)**2 + (y-0)**2)** 0.5 < epsilon:
        u = 0
        v = 0
    
    
   
    else :
        up = q*(x+0.5)/((2*np.pi) * ((x+0.5)**2 + (y-0)**2)) 
        vp = q*y/((2*np.pi) * ((x+0.5)**2 + (y-0)**2)) 
        un = -q*(x-0.5)/((2*np.pi) * ((x-0.5)**2 + (y-0)**2)) 
        vn = -q*y/((2*np.pi) * ((x-0.5)**2 + (y-0)**2)) 
        
        u= 1 + up+un
        v= vp+vn
    pressure = 1.5-0.5*(u**2+v**2)
    return pressure

```

```python
xo, yo = -0.5, -0.5
n = 50
dx = dy = 1/n
p = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        
        p[i,j] = pressure(xo + i*dx, yo+j*dy)
```

```python
plt.imshow(p,cmap='hot');
plt.colorbar();
plt.contour(p,levels=1000);

#Plot of pressure field calculated by bernoulli's theorem high pressure regions are yellow
#note that the contour lines are present are more bright & dense at high pressure region 
```

# Changed Magnitude

```python
def pressure(x,y,q=1):
    epsilon = 0.01
    q1 = 1.5
    q2=0.5
    if ((x+0.5)**2 + (y-0)**2)** 0.5 < epsilon or ((x-0.5)**2 + (y-0)**2)** 0.5 < epsilon:
        u = 0
        v = 0
    
    
   
    else :
        up = q1*(x+0.5)/((2*np.pi) * ((x+0.5)**2 + (y-0)**2)) 
        vp = q1*y/((2*np.pi) * ((x+0.5)**2 + (y-0)**2)) 
        un = -q2*(x-0.5)/((2*np.pi) * ((x-0.5)**2 + (y-0)**2)) 
        vn = -q2*y/((2*np.pi) * ((x-0.5)**2 + (y-0)**2)) 
        
        u= 1 + up+un
        v= vp+vn
    pressure = 1.5-0.5*(u**2+v**2)
    return pressure
```

```python
xo, yo = -0.5, -0.5
n = 50
dx = dy = 1/n
p = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        
        p[i,j] = pressure(xo + i*dx, yo+j*dy)
```

```python
plt.imshow(p,cmap='hot');
plt.colorbar();
plt.contour(p,levels=1000);

#Plot for different magnitude of Source (1.5) & Sink (-0.5) one can observe the smaller and less dense contour lines around sink
```
