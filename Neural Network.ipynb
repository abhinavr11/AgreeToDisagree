```python
import numpy as np
```

```python
class Model:
    
    def __init__(self, lr):
       
    #Here I define the network architecture variables
    
        self.weight_layer1 = np.random.normal(0, 0.1, 10)
        self.bias_layer1 = np.random.normal(0, 0.1, 5)
        self.weight_layer2 = np.random.normal(0, 0.1, 15)
        self.bias_layer2 = np.random.normal(0, 0.1, 3)
        self.hidden_layer = np.array([1,2,3,4,5],dtype=float)  
        self.last_layer = np.array([0.1,.2,.4],dtype=float) 
        self.preds = np.array([0.2,0.4,0.5],dtype=float) 
        self.lr = lr
        self.loss = 0 
        
        
        
   #A function just to calculate softmax

    def _softmax (self):
        
        sum_exp = 0
        
        for l in range(3):
            sum_exp += np.exp(self.last_layer[l])
            
        
        for i in range(3):
            self.preds[i] = np.exp(self.last_layer[i])/sum_exp
        
           
        
        #self.loss = self.labels.argmax()*np.log(self.preds.argmax()) + (1-self.labels.argmax())*np.log(1 - self.preds.argmax())
        print('Prediction :','Actual Label')
        print(self.preds.argmax(),':',self.labels.argmax()) 
        
  
 #Function to give back softmax while predicting
    
    def softmax_predict(self,ll):
        sum_exp =0
        predictions = np.array([1,1,1],dtype=float) 
        
        for l in range(3):
            sum_exp += np.exp(ll[l])
            
        
        for i in range(3):
            predictions[i] = np.exp(ll[i])/sum_exp
        
        return predictions
        
    def _backward(self,x11,x22):
        alpha = self.lr
        lbl = self.labels
        y = self.last_layer
        p = self.preds
        h = self.hidden_layer
        x = [x11,x22]
        
        # Updating weights of layer 1
        
        for j in range (10):
            
            for i in range(3):
                
                self.weight_layer1[j] += -1*alpha * (lbl[i]/p[i] + (1-lbl[i])/(p[i]-1)) * (p[i])*(1+p[i])*(self.weight_layer2[5*i + int(j/2)]) * h[int(j/2)] * (1-h[int(j/2)])*x[j%2]
         
        
        #updating bias of layer 1
        
        for j in range(5):
            
            for i in range(3):
                
                self.bias_layer1[j] += -alpha*(lbl[i]/p[i] + (1-lbl[i])/(p[i]-1))*(p[i])*(1+p[i])*(self.weight_layer2[5*i+j])*h[j]*(1-h[j])
        
                
        
        # Updateing weights of layer 2
        for i in range(5):
            self.weight_layer2[i] += -alpha*(lbl[0]/p[0] + (1-lbl[0])/(p[0]-1))*(p[0])*(1+p[0])*h[i]
        
        
        for i in range(5,10):
            self.weight_layer2[i] += -alpha*(lbl[1]/p[1] + (1-lbl[1])/(p[1]-1))*(p[1])*(1+p[1])*h[i-5]
            
        
        for i in range(10,15):
            self.weight_layer2[i] += -alpha*(lbl[2]/p[2] + (1-lbl[2])/(p[2]-1))*(p[2])*(1+p[2])*h[i-10]
        
        # updating bias of layer 2
        for i in range(3):
            self.bias_layer2[i] += -alpha*(lbl[i]/p[i] + (1-lbl[i])/(p[i]-1))*(p[i])*(1+p[i])
            
        
        print(f'Weights of layer 1: {self.weight_layer1}')
        print(f'Weights of layer 2: {self.weight_layer2}')
        print(f'Bias of layer 1: {self.bias_layer1}')
        print(f'Bias of layer 2: {self.bias_layer2}')
        
        
    def fit(self,data,epochs = 1):
        
        self.dataset = data  
        
        for num_epochs in range(epochs):     #Loop runs for number of epochs
            
        
            print(f'Number of epoch ->{num_epochs}')
        
            for x1, x2,x3 in zip(data[0],data[1],data[2]):
                print("___________ \n")  
            
                if x3 == 0:
                    self.labels = np.array([1,0,0])
                elif x3 == 1:
                    self.labels = np.array([0,1,0])
                else :
                    self.labels = np.array([0,0,1])
            
            #forward propogation of network
                
                wi = 0
                for i in range(5):
                
                    self.hidden_layer[i] =1/(1+np.exp(-(self.weight_layer1[wi]*x1+self.weight_layer1[wi+1]*x2+self.bias_layer1[i])))
                    wi += 2                          
          
                p = 0
                for i in range(3):
                
                
                    sum_temp = np.array([0,0,0],dtype=float)
                    for j in range(5):
                    
                    
                        sum_temp[i] += self.weight_layer2[p]*self.hidden_layer[j] 
                        p +=1
                    
                    self.last_layer[i] = self.bias_layer2[i] + sum_temp[i]                       
            
            #For caclculation of probabilities from logits
                self._softmax()
            
            #For back propogation
                self._backward(x1,x2)
            
    def predict(self,data):
        
        
            
        hidden_layer_predict = np.array([1,1,1,1,1],dtype=float)  
        last_layer_predict = np.array([1,1,1],dtype=float) 
            
        for x1 , x2 in zip(data[0],data[1]):
            
            
                
            wi = 0
            for i in range(5):
                
                
                hidden_layer_predict[i] =1/(1+np.exp(-(self.weight_layer1[wi]*x1 + self.weight_layer1[wi+1]*x2 + self.bias_layer1[i])))
                wi += 2
                    
            
            p = 0
            for i in range(3):
                
                
                sum_temp_predict = np.array([0,0,0],dtype=float)
                for j in range(5):
                    
                    
                    sum_temp_predict[i] += self.weight_layer2[p]*hidden_layer_predict[j] 
                    p +=1
                    
                last_layer_predict[i] = self.bias_layer2[i] + sum_temp_predict[i]
                #del sum_temp_predict
               
            preds = self.softmax_predict(last_layer_predict)
                
            print(preds)
        
        
```

```python
train_data = np.array([[10,2,3,4,5,6,7],[5,6,7,5,6,8,5],[1,1,0,2,1,0,0]]) #Picked any random data it makes no actual sense first two columns contain the feature and the 3rd column contain the label
```

```python
model = Model(lr = 0.1)
```

```python
model.fit(train_data,epochs=1)
```
